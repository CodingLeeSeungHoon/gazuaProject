# Upbit Quotation API 분석

## 1. 캔들 Data

### 1-1. Day candle

```python
import requests
import json
import pandas as pd
from IPython.display import display

url = "https://api.upbit.com/v1/candles/minutes/1"

# count 사용 -> 현재로부터 4분치 데이터
querystring = {"market":"KRW-BTC","count":"4"}

response = requests.request("GET", url, params=querystring)
resp = response.json()

df = pd.DataFrame(resp)
display(df)
```
